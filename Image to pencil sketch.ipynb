<h1> Image to Pencil Sketch with Python</h1>

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

<h3> Origional Image </h3>

```python
image = cv2.imread("Elephant.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#plt.figure(figsize = (8,8))
plt.imshow(image)
plt.axis("off")
plt.show()
# cv2.imshow("Original Image", image)
# cv2.waitKey(0)
```

<h3> GrayScale Image </h3>

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize = (8,8))
plt.imshow(gray_image, cmap = "gray")
plt.axis("off")
plt.show()
```

<h3> Inverting GrayScale Image </h3>

```python
inverted_image = cv2.bitwise_not(gray_image)
plt.figure(figsize = (8,8))
plt.imshow(inverted_image, cmap = "gray")
plt.axis("off")
plt.show()
```

<h3> Blurring the Inverted Image </h3>

```python
image_smoothing = cv2.GaussianBlur(inverted_image, (21,21),sigmaX = 0, sigmaY = 0)
plt.figure(figsize = (8,8))
plt.imshow(image_smoothing, cmap = "gray")
plt.axis("off")
plt.show()
```

<h3> Sketch </h3>

```python
pencil_sketch = cv2.divide(gray_image, 255 - image_smoothing ,scale = 255)
plt.figure(figsize = (8,8))
plt.imshow(pencil_sketch, cmap = "gray")
plt.axis("off")
plt.title("Final Pencil Sketch")
plt.show()
```

<h3> Comparision </h3>

```python
plt.figure(figsize = (50,50))
plt.subplot(1,5,1)
plt.imshow(image)
plt.axis("off")
plt.subplot(1,5,2)
plt.imshow(pencil_sketch, cmap = "gray")
plt.axis("off")
plt.show()
```
